Instanciação de uma população e iniciando o seu povoamento

```python
x=[]
x.append(np.outer(np.linspace(0, 10, 30), np.ones(30)))
x.append(x[0].T)
z =np.sin(x[0])*np.sqrt(x[0])*np.sin(x[1])*np.sqrt(x[1])
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x[0], x[1], z,cmap='viridis', edgecolor='none')
ax.set_title('Surface plot')
plt.show()
```

Plot da superfície

```python
plotarCN(x=x,z=z)
```

```python
p= Populacao(quantidade = N)
p.iniciar_p()
p.population.sort(reverse=True)
plotarCN(p.population,x=x,z=z)
```

Plot da superfície inicial

```python
for i in range(max_int):
  s=selecao(n1 = n1, p = p)
  mutados=mutacao(s,rho = rho, Nc = int(Nc),N=N,max = max, min = min)
  plotarCN(mutados,x=x,z=z)
  p.population=mutados
```

Plot de cada geração de cada mutação. É possível notar que conforme vão se passando as gerações, elas vão se agrupando em seus mínimos locais

```python
x=[]
x.append(np.outer(np.linspace(0, 10, 30), np.ones(30)))
x.append(x[0].T)
z =np.sin(x[0])*np.sqrt(x[0])*np.sin(x[1])*np.sqrt(x[1])
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x[0], x[1], z,cmap='viridis', edgecolor='none',alpha=0.2)
ax.set_title('Surface plot')
for ponto in mutados:
  ax.scatter(ponto.x[0],ponto.x[1],ponto.aptidao)
plt.show()
```
