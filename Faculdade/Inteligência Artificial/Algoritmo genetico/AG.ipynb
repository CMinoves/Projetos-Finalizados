A classe `Genetico` é responsável por gerar a nova geração, ela recebe como parametro a `Populacao`, as taxas de mutação de cruzamento e mutação e se utilizara elitismo:

-`selecao()`

-`crossover()`

-`mutacao(delta)`

```python
# A seleção sorteia um valor entre 0 e a soma dos fitness, 
# se o valor for menor que a soma acumulada dos fitness k-1 k sera sorteado de  --->            k indo de 1 a N
p=Populacao(quantidade=4)
p.iniciar_p()
#  def __init__(self,populacao,taxa_cross:float=0.7,taxa_mutacao:float=0.05,elitismo=False):
g=Genetico(p)
selecionados=g.selecao(mostrar=True)
selecionados
```

```python
# O crossover  que utilizamos foi o  aritmetico
#F1 = alfa*P1+(1-alfa)*P2
#F2 = alfa*P2+(1-alfa)*P1
#com o alfa aleatorio entre 0 e 1
filhos=g.crossover(selecionados,mostrar=True)
filhos
```

```python
# A mutação, quando ocorre, a amostra filha altera o valor de pelo menos umas das suas coordenadas com uma
# porcentagem delta de seu proprio valor a mais ou menos
g.mutacao(filhos,0.1,mostrar=True)
```
